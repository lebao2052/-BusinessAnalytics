```python
import pandas as pd

# Đọc tệp CSV và lưu dữ liệu vào một DataFrame
data = pd.read_csv('BIDV_BID.csv')
data = data[['Close']]

# In ra nội dung của DataFrame
print(data)
```

```python
import pandas as pd

# Tạo một DataFrame với dữ liệu

data = pd.DataFrame(data)

# In ra các thống kê mô tả của DataFrame
print(data.describe())
```

```python
import matplotlib.pyplot as plt

# Vẽ biểu đồ histogram
plt.hist(data)

# Hiển thị biểu đồ
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Vẽ biểu đồ box plot
plt.boxplot(data)

# Hiển thị biểu đồ
plt.show()
```

```python
data.plot()
```
